# Introduction

This project explores data from the California Safe Cosmetics Program (CSCP), which requires manufacturers to disclose the use of ingredients in cosmetic products that are known or suspected to cause cancer, birth defects, or other reproductive harm. The dataset includes detailed records on cosmetic and personal care products, their brands, manufacturers, and the chemicals they contain.

## Q4: Are there trends over time in reporting? #presentation

## Q6: Comparison between active and discontinuous chemicals. #presentationm

## Q7: What is the number of products per company that are discontinued? #presentation

```python
# remove the NaN to get result
discontinued_df = df[df['DiscontinuedDate'].notna() & (df['DiscontinuedDate'].astype(str).str.strip() != '')]
# Group by company and count discontinued products
discontinued_counts = discontinued_df.groupby('CompanyName').size().reset_index(name='DiscontinuedProductCount')
# to get only top 10
top_discontinued = discontinued_counts.nlargest(10, 'DiscontinuedProductCount')
#showing the visulization in chart
plt.bar(top_discontinued['CompanyName'], top_discontinued['DiscontinuedProductCount'], color= 'orange')
plt.xlabel("Company Name")
plt.ylabel("No. Discontinued Products")
plt.title("Top 10 Companies")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# This visualize represent the top 10 companies with their discontinued products .
```

```python
top_discontinued
```

## Q8: what is the percentage of products reported per company?

```python
company_counts = df['CompanyName'].value_counts()
total = company_counts.sum()
company_percentages = (company_counts / total) * 100
top_n = 10
top_companies = company_percentages.head(top_n)
plt.figure(figsize=(8, 4))
top_companies.plot(kind='barh', color='orange')
plt.xlabel('Percentage of Reported Products')
plt.ylabel('Company Name')
plt.title("Top 10 Companies by Reported Products (%)")
# Plot the top 10 companies with the highest percentage of reported products
```
